# Comandos básicos do Git
 - git clone + link = clonar repositório no computador
 - git pull = sincronizar com o git
 - git status = ver a situação das modificações
 - git add . = adicionar todas as modificações em uma pasta
 - git add + nome arquivo = adicionar uma modificação específica em uma pasta
 - git commit -m" + mensagem + " = Dar nome à pasta criada pelo add, oficializar a pasta?
 - git push = mandar todas as modificações que estão em pastas pro git
 - git diff = saber exatamente o que mudou
 - git log = ver os commits no terminal
 - ls = ver os arquivos do computador
 - ls -a = ver os arquivos do computador, até os ocultos
 - cd + nome da pasta = abrir o arquivo no terminal
 - cd .. = fechar o arquivo no terminal
 - git restore --staged + nome do arquivo = retirar uma modificação que está numa pasta do add
 - git restore + nome do arquivo = voltar uma alteração (excluir ela)
# Comandos uv 
### Abrir python
 - uv init = iniciar uv
 - uv add + nome da biblioteca = adicionar biblioteca
 - uv add jupyter notebook = add o jupyter
 - uv run jupyter notebook = abrir o jupyter
 - uv sync = sincronizar os projetos e baixar oque precisa
 - uv lock + nome do arquivo (touch.gitignore) = ignorar algo
# Comandos VI 
### Serve para criar um atalho para os comandos e para outras coisas também (no computador, não nos arquivos)
 - vi.bashrc = 
 - :q! = sair do vi sem salvar 
 - esc U = desfazer
 - i = editar escrita
 - esc = sair do modo de edição
 - o = nova linha + modo de edição
 - :w = salvar as alterações
 - cat.bashrc = ver alteração
 - alias + comando novo + = + comando antigo que quer substituir
Ex.: alias jupyter = "uv run jupyter notebook"  - se escrever jupyter no terminal, vai rodar "uv run jupyter notebook"
