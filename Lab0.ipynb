```python
x = np.random.normal(0,1,100)
z= np.random.uniform(-1,1,100)
w = np.random.exponential(2,100)

distribuicoes['Normal']["dados"] = x
distribuicoes['Uniforme']["dados"] = z
distribuicoes['Exponencial']["dados"] = w
```

Passo 4

```python
media_x = np.mean(x)
media_z = np.mean(z)
media_w = np.mean(w)
dp_x = np.std(x)
dp_z = np.std(z)
dp_w = np.std(w)
var_x = np.var(x)
var_z = np.var(z)
var_w = np.var(w)
mediana_x = np.median(x)
mediana_z = np.median(z)
mediana_w = np.median(w)
est_x = {"media":media_x,"dp":dp_x,"variancia":var_x,"mediana":mediana_x}
est_z = {"media":media_z,"dp":dp_z,"variancia":var_z,"mediana":mediana_z}
est_w = {"media":media_w,"dp":dp_w,"variancia":var_w,"mediana":mediana_w}
distribuicoes['Normal']["estatisticas"] = est_x
distribuicoes['Uniforme']["estatisticas"] = est_z
distribuicoes['Exponencial']["estatisticas"] = est_w
```

```python
def calcular_operacoes_vetores(vetor1, vetor2):
    if len(vetor1) != len(vetor2):
        print("vetores nao tem o mesmo tamanho")
        return None, None
    prod = np.dot(vetor1, vetor2)
    dist = np.linalg.norm(vetor1 - vetor2)
    return prod, dist
```

Ex 4

```python
A = np.random.randint(-5, 6, (4, 4))
print(A)

# determinante
det = np.linalg.det(A)
print("det =", det)

# verificar se é invertível
if det != 0:
    A_inv = np.linalg.inv(A)
    print(A_inv)

    # verificar A * A^-1
    I = np.dot(A, A_inv)
    print(I)
else:
    print("A não é invertível")
```

Ex 5

```python
M = np.random.rand(5, 5)

# matriz simétrica
B = (M + M.T) / 2
print(B)

# autovalores e autovetores
valores, vetores = np.linalg.eig(B)

print("Autovalores:", valores)
print("Autovetores:", vetores)

# verificar Bv = lambda v (para o primeiro)
v = vetores[:, 0]
l = valores[0]

print("Bv =", np.dot(B, v))
print("lambda*v =", l * v)
```

Ex 6

```python
A = np.random.randint(1, 10, (3, 3))
y = np.random.randint(1, 10, 3)

print(A)
print(y)

# resolver sistema
x = np.linalg.solve(A, y)
print(x)

# verificar erro
erro = np.linalg.norm(np.dot(A, x) - y)
print("erro =", erro)
```

Ex 7

```python
X = np.random.randn(100, 2)

# matriz de covariância
cov = np.cov(X, rowvar=False)
print("Covariância:", cov)

# autovalores e autovetores
val, vec = np.linalg.eig(cov)

print("Autovalores:", val)
print("Autovetores:", vec)

# gráfico dos pontos
plt.scatter(X[:,0], X[:,1])

# desenhar autovetores (direções principais)
for i in range(2):
    plt.arrow(0, 0,
              vec[0,i]*val[i],
              vec[1,i]*val[i],
              head_width=0.1)

plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Direções principais")

plt.show()
```
